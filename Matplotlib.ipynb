1. What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.


Matplotlib is a popular data visualization library in Python that is used to create high-quality 2D and 3D plots. It provides a variety of functions and tools for creating different types of charts and graphs, including line plots, scatter plots, bar charts, histograms, and many more.

5 plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: A line plot is a type of chart that displays data points connected by straight lines. It is used to show trends over time or to compare two or more variables.

Scatter plot: A scatter plot is a type of chart that displays data points as dots on a two-dimensional plane. It is used to show the relationship between two variables.

Bar chart: A bar chart is a type of chart that displays data using rectangular bars. It is used to compare the values of different categories.

Histogram: A histogram is a type of chart that displays the distribution of data using bins. It is used to show the frequency of data in each bin.

Pie chart: A pie chart is a type of chart that displays data as slices of a circle. It is used to show the proportion of each category in the data.






Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot is a type of chart used to visualize the relationship between two variables. It consists of a set of data points, each represented by a dot, plotted on a two-dimensional plane, where one variable is plotted on the x-axis and the other variable is plotted on the y-axis. The position of each dot on the plane represents the values of the two variables for that data point.

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.title("Scatter plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.scatter(x,y)
```

Q3. Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create multiple plots or subplots in a single figure. It allows you to display multiple plots on a single figure, which is useful for comparing different datasets or visualizing different aspects of the same data.

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(1, 2, 1)
plt.plot(x,y)

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([50, 20, 40, 20, 60, 70])


plt.subplot(1, 2, 2)
plt.plot(x1,y1)

```

```python
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(1, 2, 1)
plt.plot(x2,y2)

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(1, 2, 2)
plt.plot(x3,y3)
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of chart used to display and compare the values of different categories of data. It consists of a series of bars, where the height or length of each bar represents the value of a particular category. Bar plots can be vertical or horizontal, depending on the orientation of the bars.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company, profit)
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a type of chart used to represent the distribution of a set of data. It provides a visual summary of key features of the data, such as the median, quartiles, and outliers.


```python
box1 = np.random.normal(100, 10, 200)
plt.boxplot(box1)
```

```python
box2 = np.random.normal(90, 20, 200)
plt.boxplot(box2)
```
